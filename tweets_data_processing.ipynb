```python
import pandas as pd
import numpy as np
import json
```

```python
with open('/content/drive/MyDrive/tweets_data/tweets_data.json', 'r', encoding = 'utf-8') as f:
    data = json.load(f)
```

```python
data_test = data[:5]
print(data_test)
print(data_test[0].keys())
```

```python
processed_data = []
```

```python
extract_field = ['tweet_body', 'language', 'reply_counts', 'quote_counts', 'bookmark_count', 'views', 'likes' ,'retweet_counts']
extract_field_user = ["username", "created_at", "fast_followers_count", "favourites_count", "followers_count", "friends_count", "listed_count" ,
                      "media_count", "location", "statuses_count", "verified"]
```

```python
for post in data:
  item = {}

  item['tweet_id'] = np.int64(post['id'])

  for field in extract_field:
    item[field] = post[field]

  item['user_id'] = post['author']['id']

  for field in extract_field_user:
    item[field] = post['author'][field]



  processed_data.append(item)
```

```python
tweet_df = pd.DataFrame(processed_data)
tweet_df.head()
```

```python
tweet_df.drop_duplicates(subset=['tweet_id'], inplace = True, ignore_index = True)
```

```python
from datetime import datetime

date_format = '%Y-%m-%d %H:%M:%S%z'

tweet_df['created_at'] = tweet_df['created_at'].apply(lambda x: datetime.strptime(x, date_format))
```

```python
with open('/content/drive/MyDrive/tweets_data/tweets_data_processed.json', 'w') as f:
    f.write(tweet_df.to_json(orient= 'columns'))
```
